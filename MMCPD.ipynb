```python
import sympy
from sympy import *
a, b, c, x, y, z = symbols('a b c x y z')
qH, qL, QH, QL, Q_bar, pH, pL = symbols('qH qL QH QL Q_bar pH pL')
init_printing()
```

```python
# Demand Function
def get_pH(QH):
    return a - 2*QH
def get_pL(QL):
    return 1 - 2*QL
```

```python
qH = QH / 2
qL = QL / 2
```

```python
pi1 = get_pH(QH)*qH + get_pL(QL)*qL
pi1
```

```python
FOC = diff(pi1.subs(QL, Q_bar - QH), QH)
FOC
```

```python
tt = solve(FOC, QH)
```

```python
factor(tt[0])
```

```python
qH = tt[0]/2
qH
```

```python
pH = a - 2*(2*qH)
pH
```
